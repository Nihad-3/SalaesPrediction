```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.decomposition import KernelPCA
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor

# Sayısal (rəqəmsal) və kateqorial xüsusiyyətlərin seçilməsi
numeric_feature = make_column_selector(dtype_include='number')
categoric_feature = make_column_selector(dtype_include='object')

# Sayısal xüsusiyyətlər üçün pipeline
numeric_transform = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
])

# Kateqorial xüsusiyyətlər üçün pipeline
categoric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoding', OneHotEncoder(handle_unknown='ignore'))
])

# Sayısal və kateqorial xüsusiyyətlər üçün transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transform, numeric_feature),  
        ('cat', categoric_transformer, categoric_feature)  
    ]
)

# Model
model = GradientBoostingRegressor()

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])

```

```python
x = df.drop(columns=['Ümumi satış'])
y = df['Ümumi satış']
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
pipeline.fit(x_train,y_train)
```

```python
y_pred = pipeline.predict(x_test)
```

```python
df
```

```python
from sklearn.metrics import mean_squared_error,r2_score

mean_error = mean_squared_error(y_test,y_pred)
r2_scored = r2_score(y_test,y_pred)
print(mean_error)
print(r2_scored)
```
